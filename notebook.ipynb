```python
def chunk_text_with_overlap(text, chunk_size, overlap):
    """Splits a long text into chunks of a specified size with overlap.

    Args:
        text (str): The input text to be chunked.
        chunk_size (int): The desired size of each chunk (in characters).
        overlap (int): The number of characters to overlap between consecutive chunks.

    Returns:
        list: A list of text chunks.
    """
    chunks =[]
    start = 0
    while start < len(text):
        end = min(start + chunk_size, len(text))
        chunks.append(text[start:end])
        if end == len(text):
            break
        start = end - overlap
    return chunks

# Define your desired chunk size and overlap
chunk_size = 500  # Example: Split into chunks of 500 characters
overlap = 100    # Example: Overlap by 100 characters

# Assuming you have already extracted the main_story as in the previous step
main_story_chunks = chunk_text_with_overlap(main_story, chunk_size, overlap)

main_story_chunks.append(characters_section)

main_story_chunks.append(title_section)
main_story_chunks.append(summary_section)
main_story_chunks.append(idea_section)
```

```python
import pandas as pd 
df = pd.DataFrame(main_story_chunks,columns=['Text'])
```

```python
from transformers import AutoTokenizer, AutoModel

# Load the tokenizer and model
model_name = "jinaai/jina-embeddings-v3"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
import torch
import pandas as pd

def embed_fn(text):
    """Generate embeddings using jinaai/jina-embeddings-v3."""
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embedding = outputs.last_hidden_state.mean(dim=1).squeeze().tolist()
    return embedding

# Compute embeddings
df["Embeddings"] = df["Text"].apply(embed_fn)
```

```python
import numpy as np
import faiss
# Convert embeddings to a numpy array if they aren't already
embeddings = np.array(df['Embeddings'].tolist()).astype('float32')

# Get the dimension of embeddings
dimension = embeddings.shape[1]

# Create a FAISS index - IndexFlatIP for cosine similarity
index = faiss.IndexFlatIP(dimension)  # IP = Inner Product

# Before adding vectors for cosine similarity, normalize them
faiss.normalize_L2(embeddings)
index.add(embeddings)
```

```python
import google.generativeai as genai


genai.configure(api_key='YOUR_GEMINI_API_KEY')


# Function to search for similar items
def search_similar(query_text, top_k=5):
    # Get embedding for the query
    query_embedding = embed_fn(query_text)
    query_embedding = np.array([query_embedding]).astype('float32')
    
    # Normalize query vector for cosine similarity
    faiss.normalize_L2(query_embedding)
    
    # Search
    distances, indices = index.search(query_embedding, top_k)
    
    # Return results
    results = []
    for i, idx in enumerate(indices[0]):
        results.append({
            'index': idx,
            'text': df.iloc[idx]['Text'],
            'similarity_score': distances[0][i]
        })
    
    return results


def generate_response(user_prompt):
    similar_items = search_similar(user_prompt, top_k=3)
    retrieved_docs=[]
    print('Retrieved documents :')
    for item in similar_items:
        retrieved_docs.append(item['text'])
        print(f"Score: {item['similarity_score']:.4f}, Text: {item['text'][:100]}...")

    print()
        
    model2 = genai.GenerativeModel("models/gemini-1.5-flash")
    
    prompt = f"""Your're an expert books assistant ,here's the retrieved documents , please answer the question 
    of the user from the retrieved documents and if the information is missing please respond with i dont know
    
    RETRIEVED_DOCUMENTS : 
    {retrieved_docs}
    
    THE_QUESTION_OF_THE_USER
    {user_prompt}
    """
    print('-'*80)
    
    # Call `count_tokens` to get the input token count (`total_tokens`).
    print("Prompt final total tokens count: ", model2.count_tokens(main_story_chunks[0]))

    print('-'*80)
    print('Response :')
    
    # Generate a response
    response = model2.generate_content(prompt)
    
    # Print the response
    print(response.text)
```

```python
prompt = "Give me the summary of this story please"
generate_response(prompt)
```
